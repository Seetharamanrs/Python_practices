```python
# Evaluate the postfix expression.
exp = "52+83-*"
stack=[]
for i in exp:
    if i.isdigit():
        stack.append(int(i))
    else:
        val2=stack.pop()
        val1=stack.pop()
        if i=="+":
            stack.append(val2+val1)
        elif i=="-":
            stack.append(val1-val2)
        elif i=="*":
            stack.append(val2*val1)
        elif i=="/":
            stack.append(int(val1/val2))
stack[0]
```

```python
# Binary Search – Count Occurrences
a = [1,2,2,2,3,4,5]
target = 2

def first_occ(a,target):
    l=0
    r=len(a)-1
    first=-1
    while l<=r:
        mid=(l+r)//2
        if a[mid]==target:
            first=mid
            r=mid-1
        elif a[mid]<target:
            l=mid+1
        else:
            r=mid-1
    return first

def last_occ(a,target):
    l=0
    r=len(a)-1
    last=-1
    while l<=r:
        mid=(l+r)//2
        if a[mid]==target:
            last=mid
            l=mid+1
        elif a[mid]<target:
            l=mid+1
        else:
            r=mid-1 
    return last
first_idx=first_occ(a,target)
last_idx=last_occ(a,target)
        
c=last_idx-first_idx+1
print(c)
```
