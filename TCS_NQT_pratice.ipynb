```python
#Count vowels in a string
def vow(a):
    vowel='aeiou'
    c=0
    for i in a:
        if i in vowel:
            c+=1
    return c
```

```python
vow("education")
```

```python
#Reverse a string without using slicing

a='python'
def rev(a):
    c=''
    for i in range(len(a)-1,-1,-1):
        c=c+a[i] 
    return c
```

```python
#Find the largest number in a list
a=[4, 8, 2, 15, 7]
def lar_lis(a):
    lar=float('-inf')
    for i in a:
        if i>lar:
            lar=i
    return lar
```

```python
lar_lis(a)
```

```python
#Palindrome

a="madam"
is_palin=True
l=0
r=len(a)-1
while l<r:
    if a[l]==a[r]:
        l+=1
        r-=1
    else:
        is_palin=False
        break
is_palin
```

```python
#Second Largest Element

a=[10, 5, 20, 8, 15]
lar=float("-inf")
sec_lar=float("-inf")
for i in a:
    if i >lar:
        sec_lar=lar
        lar=i
    elif i<lar and i>sec_lar:
        sec_lar=i
print(sec_lar)
```

```python
#Character Frequency
a="banana"
d={}
for i in a:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
d
```
