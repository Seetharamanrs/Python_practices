```python
#Count vowels in a string
def vow(a):
    vowel='aeiou'
    c=0
    for i in a:
        if i in vowel:
            c+=1
    return c
```

```python
vow("education")
```

```python
#Reverse a string without using slicing

a='python'
def rev(a):
    c=''
    for i in range(len(a)-1,-1,-1):
        c=c+a[i] 
    return c
```

```python
#Find the largest number in a list
a=[4, 8, 2, 15, 7]
def lar_lis(a):
    lar=float('-inf')
    for i in a:
        if i>lar:
            lar=i
    return lar
```

```python
lar_lis(a)
```

```python
#Palindrome

a="madam"
is_palin=True
l=0
r=len(a)-1
while l<r:
    if a[l]==a[r]:
        l+=1
        r-=1
    else:
        is_palin=False
        break
is_palin
```

```python
#Second Largest Element

a=[10, 5, 20, 8, 15]
lar=float("-inf")
sec_lar=float("-inf")
for i in a:
    if i >lar:
        sec_lar=lar
        lar=i
    elif i<lar and i>sec_lar:
        sec_lar=i
print(sec_lar)
```

```python
#Two Sum
a=[2,7,11,15]
target=9

l=0
r=len(a)-1
pair=[]
while l<r:
    sum=a[l]+a[r]
    if sum==target:
        pair.append((a[l],a[r]))
        l+=1
        r-=1
    elif sum <target:
        l+=1
    else:
        r-=1
print(pair)
```

```python
def sum(a):
    s=0
    for i in a:
        s+=i
    return s
```

```python
#Maximum Sum of Size K
a = [2,1,5,1,3,2]
k = 3
wind=sum(a[:k])
max_sum=wind
for i in range(k,len(a)):
    wind=wind+a[i]-a[i-k]
    max_sum=max(wind,max_sum)
max_sum
```

```python
##Average of Every Window
a = [1,2,3,4,5]
k = 2
wind=sum(a[:k])
aver=[wind/k]
for i in range(k,len(a)):
    wind=wind+a[i]-a[i-k]
    aver.append(wind/k)
aver
```

```python
#Maximum Sum of Size 2
a = [4,2,1,7,8,1]
k = 2
wind=a[0]+a[1]
max_sum=wind
for i in range(k, len(a)):
    wind=wind+a[i]-a[i-k]
    max_sum=max(max_sum,wind)
max_sum
```
