```python
#Count vowels in a string
def vow(a):
    vowel='aeiou'
    c=0
    for i in a:
        if i in vowel:
            c+=1
    return c
```

```python
#Two Sum
a=[2,7,11,15]
target=9

l=0
r=len(a)-1
pair=[]
while l<r:
    sum=a[l]+a[r]
    if sum==target:
        pair.append((a[l],a[r]))
        l+=1
        r-=1
    elif sum <target:
        l+=1
    else:
        r-=1
print(pair)
```

```python
#Maximum Sum of Size K
a = [2,1,5,1,3,2]
k = 3
wind=sum(a[:k])
max_sum=wind
for i in range(k,len(a)):
    wind=wind+a[i]-a[i-k]
    max_sum=max(wind,max_sum)
max_sum
```

```python
#Two Sum using Dictionary
a = [2,7,11,15]
target = 9
d={}
for i in range(len(a)-1):
    for j in range(i+1, len(a)):
        sum =a[i]+a[j]
        if sum==target:
            d[a[i],a[j]]=sum

for k,v in d.items():
    if v==target:
        print(k)
        break
```

```python
# Find the longest consecutive sequence of identical characters
s = "aabbbbccddddd"
cou=1
max_ele=s[0]
prev=s[0]
max_len=1
for r in range(1,len(s)):
    if s[r]==prev:
        cou+=1
    elif s[r]!=prev:
        prev=s[r]
        cou=1   
    if cou>max_len:
        max_len=cou
        max_ele=s[r]
print((max_ele,max_len))
```
