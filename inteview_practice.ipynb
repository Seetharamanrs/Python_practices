```python
#GCD
a=int(input("Enter the number "))
b=int(input("Enter the number: "))
def GCD(a,b):
    if a<b:
        small=a
    else:
        small=b

    for i in range(2,small+1):
        if a%i==0 and b%i==0:
            d=i
    return d

```

```python
# Employee Department Report (Grouping + Dictionaries)
employees = [
    ("IT", "Rahul"),
    ("HR", "Priya"),
    ("IT", "Amit"),
    ("Sales", "Riya"),
    ("HR", "Karan"),
    ("IT", "Meera")
]

d={}
for dept, name in employees:
    if dept not in d:
        d[dept]=[name]
    else:
        d[dept].append(name)
d=dict(sorted(d.items(),key=lambda x:x[0]))
for dep,names in d.items():
    names.sort()
d
```
