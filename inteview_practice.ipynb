```python
#GCD
a=int(input("Enter the number "))
b=int(input("Enter the number: "))
def GCD(a,b):
    if a<b:
        small=a
    else:
        small=b

    for i in range(2,small+1):
        if a%i==0 and b%i==0:
            d=i
    return d

```

```python
# Binary Search – Count Occurrences
a= [1,2,2,2,3,4,4,4,5]
target = 4
def first_occ(a,target):
    l,r=0,len(a)-1
    first=-1
    while l<=r:
        mid=(l+r)//2
        if a[mid]==target:
            first=mid
            r=mid-1
        elif a[mid]<target:
            l=mid+1
        else:
            r=mid-1
    return first

def last_occ(a,target):
    l,r=0,len(a)-1
    last=-1
    while l<=r:
        mid=(l+r)//2
        if a[mid]==target:
            last=mid
            l=mid+1
        elif a[mid]<target:
            l=mid+1
        else:
            r=mid-1
    return last
first_idx=first_occ(a,target)
last_idx=last_occ(a,target)
if first_idx==-1:
    count=0
else:
    count=last_idx-first_idx+1
print(count)
```
