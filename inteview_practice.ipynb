```python
#GCD
a=int(input("Enter the number "))
b=int(input("Enter the number: "))
def GCD(a,b):
    if a<b:
        small=a
    else:
        small=b

    for i in range(2,small+1):
        if a%i==0 and b%i==0:
            d=i
    return d

```

```python
#LCM
def lcm(a,b):
    lcm=(a*b) // GCD(a,b)
    return lcm

lcm(12,18)
```

```python
def num_digits(a):
    b=a
    c=0
    while b>0:
        c=c+1
        b=b//10
    return c
def sum_digits(a):
    b=a
    c=0
    while b>0:
        dig=b%10
        c+=dig
        b=b//10
    return c
def revers(a):
    b=a
    n=0
    while b>0:
        dig=b%10
        n=(n*10)+dig
        b=b//10
    return n
def palindro(a):
    if revers(a)==a:
        print("It's a Palindrome Number")
    else:
        print("It's not a Palindrome Number")
def main():
    while True:
        n= int(input("1. Count the digits " \
        "2. Sum of digits "\
            "3.Reverse of number "\
                "4. Palindrome or not"\
                    " 5. Exit"))
        if n==1:
                a=int(input("Enter the number: "))
                print(num_digits(a))
        elif n==2:
            a=int(input("Enter the number: "))
            print(sum_digits(a))
        elif n==3:
            a=int(input("Enter the number: "))
            print(revers(a))
        elif n==4:
            a=int(input("Enter the number: "))
            print(palindro(a))
        elif n==5:
            print("Thank you !")
            break
        else:
            print("Please enter the correct option! ")

if __name__=="__main__":
    main()



```

```python
def prime_number(n):
    if n<=1:
        return False
    else:
        for i in range(2,int(n**0.5)+1):
            if n:
                return False
        else:
            return True
def prime_number_count(n):
    a=[]
    for i in range(2,n+1):
        if prime_number(i):
            a.append(i)
    return a


def main():
    while True:
        n=int(input( "1. Check prime"\
        "2. Print primes till N"\
        "3. Exit"))
        if n==1:
            a=int(input("Please enter the number to check: "))
            if prime_number(a):
                print("It's a Prime number")
            else:
                print("NOT a Prime number")
        elif n==2:
            a=int(input("Enter the end number: "))
            print(*prime_number_count(a))
        elif n==3:
            print("Thank you !")
            break
        else:
            print("Please enter the correct option! ")

if  __name__=="__main__":
    main()

```

```python
#Check Sorted List

def sr(a):
    if len(a)<=1:
        return True
    prev=a[0]
    for i in range(1,len(a)):
        if a[i]<prev:
            return False
        prev=a[i]
    return True
    
a = [1,2,3,4]
sr(a)
```
