```python
# COUNT EACH VOWELS IN STRING
n=input("Enter the string: ")
vowels='aeiou'
count=0
for w in n:
    if w in vowels:
        count+=1    
print("Vowels counts: ",count)
```

```python
# COUNT EACH VOWELS IN STRING
n=input("Enter the string: ")
vowels={'a':0,"e":0,"i":0,"o":0,"u":0}

for l in n:
    if l in vowels:
        vowels[l]+=1
    else:
        continue
print(vowels)
```

```python
# sum of list
n=int(input("Enter how many number of list "))
lst=[]
for i in range(0,n):
    while True:
        try:
            lst.append(int(input("Enter the list numbers")))
            break
        except ValueError:
            print("Enter the number")
sum=0
for i in lst:
    sum=sum+i
print(sum)
```

```python
#Palindrome:
n=input("Enter the string to check palindrome")
if n==n[::-1]:
    print("It's palindrome")
else:
    print("Not Palindrome")
```

```python
#FIZZBUZZ
while True:
    try:
        n= int(input("Enter the end number: "))
        break
    except ValueError:
        print("Enter the number Please")
for i in range(1,n+1):
    if i%3==0 and i%5==0:
        print("FIZZBUZZ")
    elif i%3==0:
        print("FIZZ")
    elif i%5==0:
        print("BUZZ")
    else:
        print(i)
```

```python
l=[56,13,45,7,87,31,38,2]
for i in range(0,len(l)):
    mini=i
    for j in range(i+1,len(l)):
        if (l[j]<l[mini]):
            mini=j
    l[i],l[mini]=l[mini],l[i]
```
