```python
# EVEN OR ODD

n=int(input("Enter the number"))

if n%2==0:
    print("EVEN")
else:
    print("ODD")
```

```python
# COUNT EACH VOWELS IN STRING
n=input("Enter the string: ")
vowels='aeiou'
count=0
for w in n:
    if w in vowels:
        count+=1    
print("Vowels counts: ",count)
```

```python
# COUNT EACH VOWELS IN STRING
n=input("Enter the string: ")
vowels={'a':0,"e":0,"i":0,"o":0,"u":0}

for l in n:
    if l in vowels:
        vowels[l]+=1
    else:
        continue
print(vowels)
```

```python
# sum of list
n=int(input("Enter how many number of list "))
lst=[]
for i in range(0,n):
    while True:
        try:
            lst.append(int(input("Enter the list numbers")))
            break
        except ValueError:
            print("Enter the number")
sum=0
for i in lst:
    sum=sum+i
print(sum)
```

```python
#Palindrome:
n=input("Enter the string to check palindrome")
if n==n[::-1]:
    print("It's palindrome")
else:
    print("Not Palindrome")
```
