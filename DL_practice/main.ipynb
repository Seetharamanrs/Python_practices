# DEEP LEARNING

## perceptron

```python
x_input=[0.2,0.2,0.2,0.5]
w_weight=[0.1,0.2,0.3]
thersold=0.5
def step(weighted_sum):
    if weighted_sum>thersold:
        return 1
    else:
        return 0

def perceptron():
    weighted_sum=0
    for x,w in zip(x_input,w_weight):
        weighted_sum+=x*w
        print(weighted_sum)
    return step(weighted_sum)

output=perceptron()
print(output)
```
