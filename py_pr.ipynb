```python
# Fibonacci series
n=7
j=1
d=[0]
tot=0
for i in range(n):
    temp=tot
    i=j
    j=temp
    tot=i+j
    d.append(tot)
d
```

```python
# Counting Digits , Letter, and Spaces in a string
import re
name="Python is 1"

count_dig=re.sub("[^0-9]","",name)
letter_cou=re.sub("[^a-zA-Z]","",name)
space_count=re.findall("[ \s]",name)
print(len(count_dig))
print(len(letter_cou))
print(len(space_count))
```

```python
# Variable length argument
def averge(**kwargs):
    values=list(kwargs.values())
    avg=sum(values)/len(values)
    return avg

averge(num1=0,num2=10,num3=4,num5=10)
```
