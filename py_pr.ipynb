```python
# Fibonacci series
n=7
j=1
d=[0]
tot=0
for i in range(n):
    temp=tot
    i=j
    j=temp
    tot=i+j
    d.append(tot)
d
```

```python
# Counting Digits , Letter, and Spaces in a string
import re
name="Python is 1"

count_dig=re.sub("[^0-9]","",name)
letter_cou=re.sub("[^a-zA-Z]","",name)
space_count=re.findall("[ \s]",name)
print(len(count_dig))
print(len(letter_cou))
print(len(space_count))
```

```python
# Variable length argument
def averge(**kwargs):
    values=list(kwargs.values())
    avg=sum(values)/len(values)
    return avg

averge(num1=0,num2=10,num3=4,num5=10)
```

```python
# Multi-Level Sorting (Last Name, then First Name)
a=['John Smith', 'Alice Smith', 'Bob Jones', 'Diana Prince']
a.sort(key=lambda x:(x.split()[1],x.split()[0]))
a
```

```python
# Sort by Absolute Value
a=[-10, 5, -3, 1, 7]
a.sort(key=lambda x:abs(x))
a
```

```python
# Sort Dictionary Keys by Their Values
a={'apple': 50, 'banana': 10, 'cherry': 20}
a=sorted(a.items(),key=lambda x:x[1])
res=[name for name,val in a]
res
```

```python
# Sort Strings by the Number of Vowels
a=['extravaganza', 'sky', 'queue', 'banana']
vow={"a","e","i","o","u"}
d={}
for i in a:
    c=0
    for j in i:
        if j in vow:
            c=c+1
    d[i]=c
d=sorted(d.items(),key=lambda x:x[1])
res=[word for word,cou in d]
res
```

```python
# Sorting Coordinates by Distance from Origin
a=[(4, 4), (1, 2), (0, 1), (-3, -3)]
a.sort(key=lambda x:(((x[0])**2)+(x[1]**2)))
a
```

```python
# Ignore Case and Special Characters When Sorting
a=['  banana', 'Apple', ' cherry', 'apricot']
a.sort(key=lambda x:x.strip().lower())
a
```

```python
# Group or Sort Even and Odd Numbers
a=[5, 2, 9, 1, 4, 6, 8, 3]
a.sort(key=lambda x:(x%2==0,-x),reverse=True)
a
```

```python
# Sort by Frequency of Characters
e="tree"
e=list(sorted(e,key=lambda x:(-e.count(x),x)))
e
```

```python
# Sort a List of Strings by Their Last Character
a=['grapes', 'mango', 'apple', 'banana']
a.sort(key=lambda x:x[-1])
a
```

```python
# Sort Nested Lists by the Sum of Their Elements
a=[[1, 2, 3], [4, 5], [10], [1, 1, 1]]
a.sort(key=lambda x:-sum(x))
a
```

```python
# Sort IP Addresses (Version Numbers)
a=['1.10.2', '1.2.1', '2.0.0', '1.2.10']
a.sort(key=lambda x:[int(i) for i in x.split(".")])
a
```

```python
# Sort Elements by Closest to a Target Number
a=[10, 4, 6, 8, 2]
target=7
a.sort(key=lambda x: abs(target-x))
a
```
